```python
df_kiniela=generar_resultados(jornada33,completo)

jornada_goles_local=generar_goles_local(jornada33,completo)
jornada_goles_visitante=generar_goles_visitante(jornada33,completo)

jornada_tarjetas_local=generar_tarjetas_local(jornada33,completo)
jornada_tarjetas_visitante=generar_tarjetas_visitante(jornada33,completo)

jornada_corners_local=generar_corners_local(jornada33,completo)
jornada_corners_visitante=generar_corners_visitante(jornada33,completo)
```

```python
#Resultados
```

```python
for i in range(0,10,1):
    print(f"{jornada33[i][0]} Vs {jornada33[i][1]}")
    print(f"1: {round((rnd_clas2_resultados.predict_proba(df_kiniela)[i][0])*100,2)}% X: {round((rnd_clas2_resultados.predict_proba(df_kiniela)[i][2])*100,2)}% 2: {round((rnd_clas2_resultados.predict_proba(df_kiniela)[i][1])*100,2)}%,")
    print("\t")
```

```python
#Goles
```

```python
filename2 = 'goles_local_model.pkl'
with open(filename2, 'rb') as archivo_entrada:
    rnd_clas_goles_local = pickle.load(archivo_entrada)
    
filename3 = 'goles_visitante_model.pkl'
with open(filename3, 'rb') as archivo_entrada:
    rnd_clas_goles_visitante = pickle.load(archivo_entrada)
```

```python
labels1=rnd_clas_goles_local.classes_
probs1=rnd_clas_goles_local.predict_proba(jornada_goles_local)

labels2=rnd_clas_goles_visitante.classes_
probs2 = rnd_clas_goles_visitante.predict_proba(jornada_goles_visitante)
```

```python
for i, (prob1, prob2) in enumerate(zip(probs1, probs2)):
    print(f"Goles del {jornada33[i][0]}:\t\t\tGoles del {jornada33[i][1]}:")
    for j, (label1, label2) in enumerate(zip(labels1, labels2)):
        print("     {} : {:.2f}%\t\t\t     {} : {:.2f}%".format(label1, prob1[j]*100, label2, prob2[j]*100))
    print("\t\t\t\t\t")
```

```python
filename4 = 'tarjetas_local_model.pkl'
with open(filename4, 'rb') as archivo_entrada:
    rnd_clas_tarjetas_local = pickle.load(archivo_entrada)
    
filename5 = 'tarjetas_visitante_model.pkl'
with open(filename5, 'rb') as archivo_entrada:
    rnd_clas_tarjetas_visitante = pickle.load(archivo_entrada)
```

```python
#Corners
```

```python
filename6 = 'corners_local_model.pkl'
with open(filename6, 'rb') as archivo_entrada:
    rnd_clas_corners_local = pickle.load(archivo_entrada)
    
filename7 = 'corners_visitante_model.pkl'
with open(filename7, 'rb') as archivo_entrada:
    rnd_clas_corners_visitante = pickle.load(archivo_entrada)
```

```python
labels3=rnd_clas_corners_local.classes_
probs3 = rnd_clas_corners_local.predict_proba(jornada_corners_local)

labels4=rnd_clas_corners_visitante.classes_
probs4 = rnd_clas_corners_visitante.predict_proba(jornada_corners_visitante)
```

```python
for i, (prob1, prob2) in enumerate(zip(probs3, probs4)):
    print(f"Corners para {jornada33[i][0]}:\t\t\tCorners para {jornada33[i][1]}:")
    for j, (label1, label2) in enumerate(zip(labels3, labels4)):
        print("     {} : {:.2f}%\t\t\t     {} : {:.2f}%".format(label1, prob1[j]*100, label2, prob2[j]*100))
    print("\t\t\t\t\t")
```
