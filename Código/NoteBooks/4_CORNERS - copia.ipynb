```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
corners=pd.read_csv("data/datas limpias/3/data_corners2.csv")
```

```python
corners
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
#Primero con los datos del partido
```

```python
#Local
```

```python
X=corners[['clas_local', 'clas_visitante', 'Posesión_local',
       'Posesión_visitante', 'Disparos_puerta_intento_local',
       'Disparos_puerta_buenos_local',
       'Paradas_visitante', 'Pases_cruzados_local',
       'Intercepciones_visitante', 'Despejes_visitante', 'Saque_puerta_visitante','%_Saques_esquina_local_favor','%_Saques_esquina_visitante_contra']]
Y=corners["Saques_esquina_local"]
```

```python
valores = Y
limites = [0, 3, 6, 8, 10,12,14,20]
etiquetas = ["1.(0-2)", "2.(3-5)", "3.(6-7)", "4.(8-9)","5.(10-11)","6.(12-13)","7.(14-19)"]
tramos = np.digitize(valores, limites)
valores_etiquetados = np.array([etiquetas[i-1] for i in tramos])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,valores_etiquetados,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rnd_clas = RandomForestClassifier(n_estimators=200,
                                 max_leaf_nodes=30,
                                 random_state=42)
rnd_clas.fit(X_train, Y_train)

y_pred_clas = rnd_clas.predict(X_test)
accuracy_score(Y_test, y_pred_clas)
```

```python
labels=rnd_clas.classes_
probs = rnd_clas.predict_proba(X_test)
```

```python
for i, prob in enumerate(probs):
    print("Probabilidades para la instancia {}: {}".format(i, prob))
    print("Etiquetas correspondientes: {}".format(labels))
```
