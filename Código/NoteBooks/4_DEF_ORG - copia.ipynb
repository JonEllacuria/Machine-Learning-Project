```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
completo=pd.read_csv("Data/completo_hasta_32_2.csv",sep=";")
```

```python
completo["G_esp_local"]=completo["G_esp_local"]/10
completo["G_esp_visitante"]=completo["G_esp_visitante"]/10
completo["Posesión_local"]=completo["Posesión_local"]/100
completo["Posesión_visitante"]=completo["Posesión_visitante"]/100
```

```python
completo["Pases_intento_local"]=completo["Pases_intento_local"].astype(int)
completo["Pases_%_local"]=completo["Pases_%_local"].astype(int)
completo["Pases_buenos_visitante"]=completo["Pases_buenos_visitante"].astype(int)
completo["Pases_%_visitante"]=completo["Pases_%_visitante"].astype(int)
completo["Disparos_puerta_intento_local"]=completo["Disparos_puerta_intento_local"].astype(int)
completo["Disparos_puerta_%_local"]=completo["Disparos_puerta_%_local"].astype(int)
completo["Disparos_puerta_buenos_visitante"]=completo["Disparos_puerta_buenos_visitante"].astype(int)
completo["Disparos_puerta_%_visitante"]=completo["Disparos_puerta_%_visitante"].astype(int)
completo["Paradas_%_local"]=completo["Paradas_%_local"].astype(int)
completo["Paradas_visitante"]=completo["Paradas_visitante"].astype(int)
completo["Paradas_%_visitante"]=completo["Paradas_%_visitante"].astype(int)
```

```python
completo["Pases_%_local"]=completo["Pases_%_local"]/100
completo["Pases_%_visitante"]=completo["Pases_%_visitante"]/100
completo["Disparos_puerta_%_local"]=completo["Disparos_puerta_%_local"]/100
completo["Disparos_puerta_%_visitante"]=completo["Disparos_puerta_%_visitante"]/100
completo["Paradas_%_local"]=completo["Paradas_%_local"]/100
completo["Paradas_%_visitante"]=completo["Paradas_%_visitante"]/100
```

```python
saques_esquina_local=completo.groupby("Equipo_local")["Saques_esquina_local"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
saques_esquina_visitante=completo.groupby("Equipo_visitante")["Saques_esquina_visitante"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
saques_esquina_local_cont=completo.groupby("Equipo_local")["Saques_esquina_visitante"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
saques_esquina_visitante_cont=completo.groupby("Equipo_visitante")["Saques_esquina_local"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
```

```python
completo["%_Saques_esquina_local_favor"]=completo["Equipo_local"].map(saques_esquina_local)
completo["%_Saques_esquina_visitante_favor"]=completo["Equipo_visitante"].map(saques_esquina_visitante)
completo["%_Saques_esquina_local_contra"]=completo["Equipo_local"].map(saques_esquina_local_cont)
completo["%_Saques_esquina_visitante_contra"]=completo["Equipo_visitante"].map(saques_esquina_visitante_cont)
```

```python
arbitro_faltas_local=completo.groupby("Árbitro")["Faltas_local"].mean().sort_values(ascending=False)
arbitro_faltas_visitante=completo.groupby("Árbitro")["Faltas_visitante"].mean().sort_values(ascending=False)
arbitro_amarillas_local=completo.groupby("Árbitro")["Amarillas_local"].mean().sort_values(ascending=False)
arbitro_amarillas_visitante=completo.groupby("Árbitro")["Amarillas_visitante"].mean().sort_values(ascending=False)
arbitro_doble_amarilla_local=completo.groupby("Árbitro")["expulsion_doble_amarilla_local"].mean().sort_values(ascending=False)
arbitro_doble_amarilla_visitante=completo.groupby("Árbitro")["expulsion_doble_amarilla_visitante"].mean().sort_values(ascending=False)
arbitro_roja_directa_local=completo.groupby("Árbitro")["expulsión_roja_directa_local"].mean().sort_values(ascending=False)
arbitro_roja_directa_visitante=completo.groupby("Árbitro")["expulsión_roja_directa_visitante"].mean().sort_values(ascending=False)
```

```python
completo["Árbitro_faltas_local"]=completo["Árbitro"].map(arbitro_faltas_local)
completo["Árbitro_faltas_visitante"]=completo["Árbitro"].map(arbitro_faltas_visitante)
completo["Árbitro_amarillas_local"]=completo["Árbitro"].map(arbitro_amarillas_local)
completo["Árbitro_amarillas_visitante"]=completo["Árbitro"].map(arbitro_amarillas_visitante)
completo["Árbitro_doble_amarilla_local"]=completo["Árbitro"].map(arbitro_doble_amarilla_local)
completo["Árbitro_doble_amarilla_visitante"]=completo["Árbitro"].map(arbitro_doble_amarilla_visitante)
completo["Árbitro_roja_directa_local"]=completo["Árbitro"].map(arbitro_roja_directa_local)
completo["Árbitro_roja_directa_visitante"]=completo["Árbitro"].map(arbitro_roja_directa_visitante)
```

```python
completo["tarjetas_total_local"]=completo["Amarillas_local"]+completo["expulsion_doble_amarilla_local"]+completo["expulsión_roja_directa_local"]
completo["tarjetas_total_visitante"]=completo["Amarillas_visitante"]+completo["expulsion_doble_amarilla_visitante"]+completo["expulsión_roja_directa_visitante"]

sum_tarjetas_local=completo.groupby("Árbitro")["tarjetas_total_local"].mean().sort_values(ascending=False)
sum_tarjetas_visitante=completo.groupby("Árbitro")["tarjetas_total_visitante"].mean().sort_values(ascending=False)

completo["Árbitro_sum_tarjetas_local"]=completo["Árbitro"].map(sum_tarjetas_local)
completo["Árbitro_sum_tarjetas_visitante"]=completo["Árbitro"].map(sum_tarjetas_visitante)
```

```python
faltas_cometidas_local=completo.groupby("Equipo_local")["Faltas_local"].mean().sort_values(ascending=False)
faltas_cometidas_visitante=completo.groupby("Equipo_visitante")["Faltas_visitante"].mean().sort_values(ascending=False)
faltas_provocadas_local=completo.groupby("Equipo_local")["Faltas_visitante"].mean().sort_values(ascending=False)
faltas_provocadas_visitante=completo.groupby("Equipo_visitante")["Faltas_local"].mean().sort_values(ascending=False)

amarillas_cometidas_local=completo.groupby("Equipo_local")["Amarillas_local"].mean().sort_values(ascending=False)
amarillas_cometidas_visitante=completo.groupby("Equipo_visitante")["Amarillas_visitante"].mean().sort_values(ascending=False)
amarillas_provocadas_local=completo.groupby("Equipo_local")["Amarillas_visitante"].mean().sort_values(ascending=False)
amarillas_provocadas_visitante=completo.groupby("Equipo_visitante")["Amarillas_local"].mean().sort_values(ascending=False)

dobles_amarillas_cometidas_local=completo.groupby("Equipo_local")["expulsion_doble_amarilla_local"].mean().sort_values(ascending=False)
dobles_amarillas_cometidas_visitante=completo.groupby("Equipo_visitante")["expulsion_doble_amarilla_visitante"].mean().sort_values(ascending=False)
dobles_amarillas_provocadas_local=completo.groupby("Equipo_local")["expulsion_doble_amarilla_visitante"].mean().sort_values(ascending=False)
dobles_amarillas_provocadas_visitante=completo.groupby("Equipo_visitante")["expulsion_doble_amarilla_local"].mean().sort_values(ascending=False)

rojas_directas_cometidas_local=completo.groupby("Equipo_local")["expulsión_roja_directa_local"].mean().sort_values(ascending=False)
rojas_directas_cometidas_visitante=completo.groupby("Equipo_visitante")["expulsión_roja_directa_visitante"].mean().sort_values(ascending=False)
rojas_directas_provocadas_local=completo.groupby("Equipo_local")["expulsión_roja_directa_visitante"].mean().sort_values(ascending=False)
rojas_directas_provocadas_visitante=completo.groupby("Equipo_visitante")["expulsión_roja_directa_local"].mean().sort_values(ascending=False)
```

```python
completo["faltas_cometidas_local"]=completo["Equipo_local"].map(faltas_cometidas_local)
completo["faltas_cometidas_visitante"]=completo["Equipo_visitante"].map(faltas_cometidas_visitante)
completo["faltas_provocadas_local"]=completo["Equipo_local"].map(faltas_provocadas_local)
completo["faltas_provocadas_visitante"]=completo["Equipo_visitante"].map(faltas_provocadas_visitante)

completo["amarillas_cometidas_local"]=completo["Equipo_local"].map(amarillas_cometidas_local)
completo["amarillas_cometidas_visitante"]=completo["Equipo_visitante"].map(amarillas_cometidas_visitante)
completo["amarillas_provocadas_local"]=completo["Equipo_local"].map(amarillas_provocadas_local)
completo["amarillas_provocadas_visitante"]=completo["Equipo_visitante"].map(amarillas_provocadas_visitante)

completo["dobles_amarillas_cometidas_local"]=completo["Equipo_local"].map(dobles_amarillas_cometidas_local)
completo["dobles_amarillas_cometidas_visitante"]=completo["Equipo_visitante"].map(dobles_amarillas_cometidas_visitante)
completo["dobles_amarillas_provocadas_local"]=completo["Equipo_local"].map(dobles_amarillas_provocadas_local)
completo["dobles_amarillas_provocadas_visitante"]=completo["Equipo_visitante"].map(dobles_amarillas_provocadas_visitante)

completo["rojas_directas_cometida_local"]=completo["Equipo_local"].map(rojas_directas_cometidas_local)
completo["rojas_directas_cometida_visitante"]=completo["Equipo_visitante"].map(rojas_directas_cometidas_visitante)
completo["rojas_directas_provocadas_local"]=completo["Equipo_local"].map(rojas_directas_provocadas_local)
completo["rojas_directas_provocadas_visitante"]=completo["Equipo_visitante"].map(rojas_directas_provocadas_visitante)
```

```python
goles_marcados_local=completo.groupby("Equipo_local")["Goles_local"].mean().sort_values(ascending=False)
goles_marcados_visitante=completo.groupby("Equipo_visitante")["Goles_visitante"].mean().sort_values(ascending=False)
goles_recibidos_local=completo.groupby("Equipo_local")["Goles_visitante"].mean().sort_values(ascending=False)
goles_recibidos_visitante=completo.groupby("Equipo_visitante")["Goles_local"].mean().sort_values(ascending=False)
```

```python
completo["goles_marcados_local"]=completo["Equipo_local"].map(goles_marcados_local)
completo["goles_marcados_visitante"]=completo["Equipo_visitante"].map(goles_marcados_visitante)
completo["goles_recibidos_local"]=completo["Equipo_local"].map(goles_recibidos_local)
completo["goles_recibidos_visitante"]=completo["Equipo_visitante"].map(goles_recibidos_visitante)
```

```python
goles_esperados_local=completo.groupby("Equipo_local")["G_esp_local"].mean().sort_values(ascending=False)
goles_esperados_visitante=completo.groupby("Equipo_visitante")["G_esp_visitante"].mean().sort_values(ascending=False)
goles_esperados_contra_local=completo.groupby("Equipo_local")["G_esp_visitante"].mean().sort_values(ascending=False)
goles_esperados_contra_visitante=completo.groupby("Equipo_visitante")["G_esp_local"].mean().sort_values(ascending=False)
```

```python
completo["goles_esperados_local"]=completo["Equipo_local"].map(goles_esperados_local)
completo["goles_esperados_visitante"]=completo["Equipo_visitante"].map(goles_esperados_visitante)
completo["goles_esperados_contra_local"]=completo["Equipo_local"].map(goles_esperados_contra_local)
completo["goles_esperados_contra_visitante"]=completo["Equipo_visitante"].map(goles_esperados_contra_visitante)
```

```python
posesion_media_local=completo.groupby("Equipo_local")["Posesión_local"].mean().sort_values(ascending=False)
posesion_media_visitante=completo.groupby("Equipo_visitante")["Posesión_visitante"].mean().sort_values(ascending=False)
```

```python
completo["posesion_media_local"]=completo["Equipo_local"].map(posesion_media_local)
completo["posesion_media_visitante"]=completo["Equipo_visitante"].map(posesion_media_visitante)
```

```python
disparos_puerta_hechos_local=completo.groupby("Equipo_local")["Disparos_puerta_buenos_local"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
disparos_puerta_hechos_visitante=completo.groupby("Equipo_visitante")["Disparos_puerta_buenos_visitante"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
disparos_puerta_recibidos_local_cont=completo.groupby("Equipo_local")["Disparos_puerta_buenos_visitante"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
disparos_puerta_recibidos_visitante_cont=completo.groupby("Equipo_visitante")["Disparos_puerta_buenos_local"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
```

```python
completo["%_disparos_puerta_hechos_local"]=completo["Equipo_local"].map(disparos_puerta_hechos_local)
completo["%_disparos_puerta_hechos_visitante"]=completo["Equipo_visitante"].map(disparos_puerta_hechos_visitante)
completo["%_disparos_puerta_recibidos_local"]=completo["Equipo_local"].map(disparos_puerta_recibidos_local_cont)
completo["%_disparos_puerta_recibidos_visitante"]=completo["Equipo_visitante"].map(disparos_puerta_recibidos_visitante_cont)
```

```python
pases_cruzados_hechos_local=completo.groupby("Equipo_local")["Pases_cruzados_local"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
pases_cruzados_hechos_visitante=completo.groupby("Equipo_visitante")["Pases_cruzados_visitante"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
pases_cruzados_recibidos_local_cont=completo.groupby("Equipo_local")["Pases_cruzados_visitante"].sum()/completo.groupby("Equipo_local")["Equipo_local"].count()
pases_cruzados_recibidos_visitante_cont=completo.groupby("Equipo_visitante")["Pases_cruzados_local"].sum()/completo.groupby("Equipo_visitante")["Equipo_visitante"].count()
```

```python
completo["%_pases_cruzados_hechos_local"]=completo["Equipo_local"].map(pases_cruzados_hechos_local)
completo["%_pases_cruzados_hechos_visitante"]=completo["Equipo_visitante"].map(pases_cruzados_hechos_visitante)
completo["%_pases_cruzados_recibidos_local"]=completo["Equipo_local"].map(pases_cruzados_recibidos_local_cont)
completo["%_pases_cruzados_recibidos_visitante"]=completo["Equipo_visitante"].map(pases_cruzados_recibidos_visitante_cont)
```

```python
completo["tarjetas_total_local"]=completo["Amarillas_local"]+completo["expulsion_doble_amarilla_local"]+completo["expulsión_roja_directa_local"]
completo["tarjetas_total_visitante"]=completo["Amarillas_visitante"]+completo["expulsion_doble_amarilla_visitante"]+completo["expulsión_roja_directa_visitante"]
```

```python
completo["tarjetas_cometidas_local"]=completo["amarillas_cometidas_local"]+completo["dobles_amarillas_cometidas_local"]+completo["rojas_directas_cometida_local"]
completo["tarjetas_cometidas_visitante"]=completo["amarillas_cometidas_visitante"]+completo["dobles_amarillas_cometidas_visitante"]+completo["rojas_directas_cometida_visitante"]
completo["tarjetas_provocadas_local"]=completo["amarillas_provocadas_local"]+completo["dobles_amarillas_provocadas_local"]+completo["rojas_directas_provocadas_local"]
completo["tarjetas_provocadas_visitante"]=completo["amarillas_provocadas_visitante"]+completo["dobles_amarillas_provocadas_visitante"]+completo["rojas_directas_provocadas_visitante"]
```
